```python
#OOP Programming:

class Solution:
    def sort012(self, arr):
        front = 0
        back = len(arr) - 1
        current = 0

        while current <= back:
            if arr[current] == 0:
                arr[front], arr[current] = arr[current], arr[front]
                current += 1
                front += 1
    
            elif arr[current] == 1:
                current += 1
    
            else:
                arr[current], arr[back] = arr[back], arr[current]
                back -= 1
```
