```python
#Functional Programming: Binary Search

def binarySearch(arr, low, high, x):

    if high >= low:

        mid = low + (high - low) // 2

        if arr[mid] == x:
            return mid

        elif arr[mid] > x:
            return binarySearch(arr, low, mid-1, x)

        else:
            return binarySearch(arr, mid + 1, high, x)

    else:
        return -1

if __name__ == '__main__':
    arr = [2, 3, 4, 10, 40]
    x = 10
    
    result = binarySearch(arr, 0, len(arr)-1, x)
    
    if result != -1:
        print("Element is present at index", result)
    else:
        print("Element is not present in array")
```

```python
#OOP Programming: Number Sorting
#Working alonside the "Sort an Array of 0s, 1s, and 2s" documentation on G4G site, to better understand the problem and work through
#the solution.


def sort012(arr):
    number = len(arr)

    low = 0
    mid = 0
    high = number - 1

    while mid <= high: 
        if arr[mid] == 0:

            arr[low], arr[mid] = arr[mid], arr[low]
            low += 1
            mid += 1

        elif arr[mid] == 1:
            mid += 1
        
        else:
            arr[mid], arr[high] = arr[high], arr[mid]
            high -= 1

if __name__ == "__main__":
    arr = [0, 1, 2, 0, 1, 2]
    number = len(arr)

    sort012(arr)

    for i in range(number):
        print(arr[i], end=' ')

```
