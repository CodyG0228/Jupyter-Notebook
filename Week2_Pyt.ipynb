```python
#4.1
secret = 4
guess = int(input("The secret is a number between 1 and 10. Enter your guess: "))
if guess == secret:
    print("Just right (correct)")
elif guess > secret:
    print("Too high. Guess again.")
else:
    print("Too low. Guess again.")

```

```python
#4.2
small = True
green = False
if small:
    if green:
        print("Pea")
    else:
        print("Cherry")
else:
    if green:
        print("Watermelon")
    else:
        print("Pumpkin")
```

```python
#6.2
guess_me = 7
number = 1

while True:
    if number < guess_me:
        print('Too low')
    elif number > guess_me:
        print('Oops')
        break
    else:
        print('Found it!')
        break
    number += 1

```

```python
#6.3
guess_me = 5
for number in range(10):
    if number < guess_me:
        print('too low')
    elif number == guess_me:
        print('found it!')
        break
    else:
        print('oops')
        break
```
