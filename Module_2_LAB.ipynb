```python
#Author: CODY GUNTER
#Application Name: Dean's vs Honor Roll
#Description: This application will take a student name and their GPA, and output whether they made the Dean's list or Honor Roll.

while True:
    last_name = input('Please enter your last name: ')
    if last_name == 'ZZZ':
        print('Exiting application.')
        break
    
    first_name = input('Please enter your first name: ')
    gpa = float(input('Please enter your GPA: '))

    if gpa >= 3.5:
        print(f"Congratulations, {first_name} {last_name}! You've made the Dean's List!")
    elif gpa >= 3.25:
        print(f"Congratulations, {first_name} {last_name}! You've made the Honor Roll!")
    else:
        print(f'Sorry, {first_name} {last_name}. Better luck next time.')
```
