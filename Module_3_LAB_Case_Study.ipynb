```python
class Vehicle:
    def __init__(self, vehicle_type):
        self.vehicle_type = vehicle_type
    
class Automobile(Vehicle):
    def __init__(self, vehicle_type, year, make, model, doors, roof):
        super().__init__(vehicle_type)
        self.year = year
        self.make = make
        self.model = model
        self.doors = doors
        self.roof = roof

print("Lab 3 Case Study")
user_vehicle_type = input("What is the type of your vehicle? > ")
user_year = input("What is the year of your vehicle? > ")
user_make = input("What is the make of your vehicle? > ")
user_model = input("What is the model of your car? > ")
user_doors = input("How many doors does your vehicle have? > ")
user_roof = input("What kind of roof does your vehicle have? > ")

my_car = Automobile(user_vehicle_type, user_year, user_make, user_model, user_doors, user_roof)

print("-----YOUR VEHICLE SPECS-----")
print(f"Vehicle Type: {my_car.vehicle_type}")
print(f"Vehicle Year: {my_car.year}")
print(f"Vehicle Model: {my_car.model}")
print(f"Amount of Doors: {my_car.doors}")
print(f"Roof Type: {my_car.roof}")
```
